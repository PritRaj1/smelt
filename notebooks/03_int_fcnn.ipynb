# 03: Int matmul and activation

No float multiply in the forward pass.

**BitNet.cpp** ([source](https://github.com/microsoft/BitNet/blob/main/src/ggml-bitnet-lut.cpp)):
activations dequantized to float after GEMM, norm + activation in float, requantized to int8 for next GEMM.

`int8 -> GEMM (int32) -> float rescale -> float SwiGLU -> float RMSNorm -> int8 quantize -> next`

**smelt**: PLAC activation stays in int32, one right-shift to int8 for next layer.

`int8 -> GEMM (int32) -> PLAC (int32 LUT) -> right-shift to int8 -> next`

## Ternary GEMM: no-multiply

Matmul: $y_j = \sum_i w_{ji} \cdot x_i$ (multiply-accumulate)

Ternary weights $w \in \{-1, 0, +1\}$:

$$y_j = \sum_{i: w=+1} x_i - \sum_{i: w=-1} x_i$$

Additions and subtractions only, the positions are known at compile time (as bitmasks).

Activations quantized to int8 (per-token absmax, 256 levels). Accumulator is int32
because summing 4096 int8 values can reach ~500K, which overflows int8 and int16.

````python
x = np.array([10, -3, 7, 0, 5, -2, 8, 1], dtype=np.int8)
w = np.array([1, -1, 0, 1, -1, 0, 1, 1], dtype=np.int8)

# standard: 8 multiplies + 7 adds
y_std = np.sum(x * w)

# ternary: 0 multiplies, just conditional adds
y_add = np.sum(x[w == 1]) - np.sum(x[w == -1])

print(f"standard mul-acc: {y_std}")
print(f"ternary add/sub: {y_add}")
print(f"match: {y_std == y_add}")
````

## TL1 lookup table

Processing weights one at a time means one conditional add/sub per weight. At 4096 weights per row, that's 4096 branches, so very slow.

TL1: process weights in pairs. For a pair (w0, w1) and activations (a0, a1), there are only 9 possible results. Precompute all 9 into a table. Then each column just looks up which result applies.

One table lookup replaces two conditional add/subs.

| index | w0 | w1 | LUT entry |
|:-----:|:--:|:--:|:----------|
| 0     | -1 | -1 | -a0 - a1  |
| 1     | -1 |  0 | -a0       |
| 2     | -1 | +1 | -a0 + a1  |
| 3     |  0 | -1 | -a1       |
| 4     |  0 |  0 | 0         |
| 5     |  0 | +1 | a1        |
| 6     | +1 | -1 | a0 - a1   |
| 7     | +1 |  0 | a0        |
| 8     | +1 | +1 | a0 + a1   |

Index = (w0+1)*3 + (w1+1). Two indices per byte (nibbles).

The BitNet solution: every column needs a lookup from the same 9-entry table (same activation pair,
different weights). `vpshufb` does 16 lookups from a 16-byte table in one cycle.
So 16 columns are processed per instruction instead of one.

Weight layout transposed: [k_pairs, n_cols] so that all columns' indices for the same
pair are contiguous in memory for the SIMD load.

Based on [BitNet.cpp](https://github.com/microsoft/BitNet/blob/main/utils/codegen_tl1.py) TL1
and [T-MAC](https://github.com/microsoft/T-MAC).

````python
# TL1 encoding
w_pairs = np.array([[1, -1], [0, 1], [-1, -1], [1, 1]], dtype=np.int8)
for w0, w1 in w_pairs:
    idx = (w0 + 1) * 3 + (w1 + 1)
    print(f"  ({w0:+d}, {w1:+d}) -> index {idx}")
````

## GEMM C kernel

MR=2 register microkernel: processes row pairs so each weight chunk is loaded only once, but applied to two different LUTs. Saves weight memory bandwidth for batch/prefill, falls back to single row for odd m:

```
For each k-pair p:
  LUT_0 = build_lut(row0_activations[p*2], row0_activations[p*2+1])
  LUT_1 = build_lut(row1_activations[p*2], row1_activations[p*2+1])

  For each column chunk jb (32 bytes = 64 columns):
    weight_indices = load(w_tl1[p, jb:jb+32])    <- ONE load, used twice
    acc_row0[jb..] += vpshufb(LUT_0, weight_indices)
    acc_row1[jb..] += vpshufb(LUT_1, weight_indices)
```

Per activation pair (a0, a1):
1. Build 9-entry int16 LUT of all partial sums
2. Split LUT into two byte tables (lo_byte, hi_byte) because vpshufb operates on bytes, not int16. Two parallel lookups, then `unpacklo/hi_epi8` recombines into int16.
3. Broadcast 16-byte table into both 128-bit lanes (vpshufb cannot cross lanes)
4. Per 32 weight bytes: extract lo nibble (AND 0x0F) and hi nibble (shift >> 4, AND 0x0F) as even/odd column indices. vpshufb lookup, unpack to int16, accumulate.
5. Flush int16 -> int32 every 128 pairs. Overflow budget: worst case per pair is 127+127 = 254, so 128 pairs reaches ~32K (int16 max).

```
LUT entry = 0x01_23 (int16)
              |
    hi_table[i] = 0x01    lo_table[i] = 0x23

    vpshufb(lo_table, idx) -> lo_bytes
    vpshufb(hi_table, idx) -> hi_bytes
    unpacklo(lo, hi)       -> int16 results
```

| instruction | what it does |
|:---|:---|
| `_mm256_broadcastsi128_si256` | duplicate 16-byte LUT into both 128-bit lanes |
| `_mm256_shuffle_epi8` | 32 parallel byte lookups (16 per lane, independent) |
| `_mm256_unpacklo/hi_epi8` | interleave lo/hi bytes back into int16 |
| `_mm256_cvtepi16_epi32` | widen int16 to int32 during flush |

## Q16.16 fixed-point

CPUs do integer math in 1 cycle, float in 3-5. But we need fractions (0.75, 1.2, etc).

Q16.16: multiply everything by $2^{16}$ = 65536 and work in int32. 16 bits integer, 16 bits fraction.

| value | Q16.16 |
|:------|-------:|
| 1.0   | 65536  |
| 0.5   | 32768  |
| 1.5   | 98304  |
| -0.25 | -16384 |

When you multiply two Q16.16 numbers, the scale doubles ($65536^2$).
Shift right by 16 to correct: `(a * b) >> 16`.

This is how PLAC, softmax, norms, and RoPE all avoid float arithmetic.

## PLAC: replacing piecewise-linear multiplies with shifts

A piecewise linear (PWL) activation is:
$$f(x) \approx m_i \cdot x + b_i \text{ on segment } [x_i, x_{i+1}]$$

Normal PWL needs a float multiply ($m_i \cdot x$).

ML-PLAC constrains slopes to sums of signed powers of 2:
$$m_i = \sum_k s_k \cdot 2^{e_k}$$

Then $m_i \cdot x$ becomes shifts and adds:
$$0.75 \cdot x = (2^0 - 2^{-2}) \cdot x = x - (x \gg 2)$$

````python
from smelt.plac import quantize_slope, terms_to_str

slopes = [0.75, 0.42, 0.9, -0.18]
print(f"{'slope':>8s}  {'shift expression':>24s}  {'approx':>8s}  {'error':>8s}")
for s in slopes:
    terms, val = quantize_slope(s, n_terms=2)
    print(f"{s:>8.3f}  {terms_to_str(terms):>24s}  {val:>8.4f}  {abs(s - val):>8.4f}")
````

## PLAC C kernel: dense LUT

At init, the PLAC segments are baked into a dense 4096-entry int32 table.
Runtime is one subtract + shift to get the table index. No segment search, no branches.

```
y = lut[(x - x_lo) >> shift]
```

The LUT fits in L1 cache (16KB). Evaluation is memory-bound, not compute-bound.

````python
from smelt.plac import from_fixed, to_fixed

# 0.75 * 3.5 using only shifts
x = to_fixed(np.array([3.5]))[0]

# 0.75 = 1 - 0.25 = 2^0 - 2^(-2)
y = x - (x >> 2)  # shift right by 2 = divide by 4

print(f"x fixed:       {x} (= {from_fixed(np.array([x]))[0]:.4f})")
print(f"x - (x >> 2):  {y} (= {from_fixed(np.array([y]))[0]:.4f})")
print(f"expected:       {0.75 * 3.5:.4f}")
````

## FCNN layer: ternary GEMM + PLAC SiLU

````python
import torch.nn as nn

from smelt.matmul import TernaryLinear
from smelt.plac import PLACFunc


def silu(x):
    return torch.nn.functional.silu(x.float()).to(x.dtype)


torch.manual_seed(0)
linear = nn.Linear(256, 128, bias=False)
x = torch.randn(16, 256)

# float reference
y_ref = torch.nn.functional.silu(linear(x))

# smelt pipeline
tl = TernaryLinear(linear)
plac_silu = PLACFunc(silu, -8, 8, target_mae=1e-2)
y_smelt = plac_silu(tl(x))

nmse = ((y_ref - y_smelt) ** 2).mean() / (y_ref**2).mean()
print(f"NMSE (ternary GEMM + PLAC SiLU vs float): {nmse:.4f}")
````

````python
fig, ax = plt.subplots(figsize=(6, 4))
ax.scatter(
    y_ref.detach().flatten().numpy(),
    y_smelt.detach().flatten().numpy(),
    s=1,
    alpha=0.3,
)
lim = max(y_ref.abs().max().item(), y_smelt.abs().max().item())
ax.plot([-lim, lim], [-lim, lim], "k--", lw=0.5)
ax.set_xlabel("float (linear + SiLU)")
ax.set_ylabel("smelt (ternary GEMM + PLAC)")
ax.set_title(f"end-to-end NMSE={nmse:.4f}")
plt.tight_layout()
plt.show()
````

## Benchmark: GEMM latency

smelt ternary (TL1 + vpshufb + OpenMP) vs torch float32 CPU vs GPU.

````python
import platform
import subprocess
import time


def get_cpu_name():
    try:
        out = subprocess.check_output(["lscpu"], text=True)
        for line in out.splitlines():
            if "Model name" in line:
                return line.split(":")[1].strip()
    except Exception:
        pass
    return platform.processor() or platform.machine()


cpu_name = get_cpu_name()
gpu_name = torch.cuda.get_device_name(0) if torch.cuda.is_available() else None
print(f"CPU: {cpu_name}")
if gpu_name:
    print(f"GPU: {gpu_name}")

results = []
for batch, label in [(1, "batch=1"), (32, "batch=32")]:
    for k in [1024, 4096]:
        linear_layer = nn.Linear(k, k, bias=False)
        tl = TernaryLinear(linear_layer)
        xb = torch.randn(batch, k)

        linear_layer(xb)
        t0 = time.perf_counter()
        for _ in range(100):
            linear_layer(xb)
        t_cpu = (time.perf_counter() - t0) / 100

        tl(xb)
        t0 = time.perf_counter()
        for _ in range(100):
            tl(xb)
        t_smelt = (time.perf_counter() - t0) / 100

        t_gpu = None
        if torch.cuda.is_available():
            lg = linear_layer.cuda()
            xg = xb.cuda()
            for _ in range(10):
                lg(xg)
            torch.cuda.synchronize()
            t0 = time.perf_counter()
            for _ in range(100):
                lg(xg)
            torch.cuda.synchronize()
            t_gpu = (time.perf_counter() - t0) / 100

        results.append(
            {
                "name": f"{k} {label}",
                "cpu": t_cpu * 1e3,
                "smelt": t_smelt * 1e3,
                "gpu": t_gpu * 1e3 if t_gpu else None,
            }
        )

names = [r["name"] for r in results]
has_gpu = any(r["gpu"] is not None for r in results)
x_pos = np.arange(len(names))
w = 0.8 / (3 if has_gpu else 2)

fig, ax = plt.subplots(figsize=(10, 4))
cpu_label = f"torch CPU ({cpu_name})"
ax.bar(x_pos - w, [r["cpu"] for r in results], w, label=cpu_label, color="steelblue")
ax.bar(x_pos, [r["smelt"] for r in results], w, label="smelt ternary (CPU)", color="coral")
if has_gpu:
    ax.bar(
        x_pos + w,
        [r["gpu"] or 0 for r in results],
        w,
        label=f"torch GPU ({gpu_name})",
        color="seagreen",
    )
ax.set_xticks(x_pos)
ax.set_xticklabels(names)
ax.set_ylabel("ms")
ax.legend(fontsize=8)
ax.set_title("GEMM latency")
plt.tight_layout()
plt.show()

for r in results:
    ratio = r["cpu"] / r["smelt"]
    tag = f"{ratio:.1f}x faster" if ratio > 1 else f"{1 / ratio:.1f}x slower"
    line = f"  {r['name']:>16s}: cpu={r['cpu']:.2f}ms  smelt={r['smelt']:.2f}ms  {tag}"
    if r["gpu"] is not None:
        line += f"  gpu={r['gpu']:.2f}ms"
    print(line)
````

## FPGA path

smelt uses only: integer add/sub, bit-shifts, comparisons, table lookups.
No FPU, no multiplier in the pipeline.

Dedicated hardware to leverage this would need something like:

- **Wide integer ALU** (32+ lanes of int8 add/sub) for ternary GEMM acc
- **Byte shuffle unit** (vpshufb equivalent) for TL1 LUT lookups
- **Barrel shifter** for PLAC slope (variable-width shifts)
- **Small SRAM LUT** (~16KB) for dense PLAC activation tables

Existing CPUs with AVX2/AVX-512 already have the first two. FPGA/ASIC could drop float support entirely, saving huge die area.
NPUs (e.g. Apple ANE, Qualcomm Hexagon) have fixed-point datapaths that
map fine to this pipeline.
